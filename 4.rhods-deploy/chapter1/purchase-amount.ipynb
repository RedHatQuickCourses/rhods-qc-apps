# About Training a Model
In the following cells we will train a simple model, so that we can export it, and use it in an inference engine.

```python
# Include required packages for execution
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import pickle
```

```python
# Create data
numpy.random.seed(2)
# Create a random set of numbers for a mock representation of minutes spent online
x = numpy.random.normal(3,1,100)
# Create a random set of number for a mock representation of the purchase amount
y = numpy.random.normal(150,40,100) / x
```

```python
 # Create a graph with the data
# Imagine X as the number of minutes to make a purchase, and Y the purchase amount
plt.scatter(x,y)
plt.show()
```

```python
# Divide the data between tranin and test
# where the training set is the 80% of the data
# while the test set is the remaining 20%
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

# Display the training set
plt.scatter(train_x, train_y)
plt.show()
```

```python
# display the test set
plt.scatter(test_x, test_y)
plt.show()
```

```python
# train a model using polynomial regression
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

# define a line to represent the model in a plot
myline = numpy.linspace(0, 6, 100)

# plot the training set with the model line
plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
# Print R2 (R-Square), which evaluates the relationship between axis for training set 
r2_train = r2_score(train_y, mymodel(train_x))
print(r2_train)

# Print R2 (R-Square), which evaluates the relationship between axis for test set
r2_test = r2_score(test_y, mymodel(test_x))
print(r2_test)

# predict a value
print(mymodel(5))

# export the model to a file
filename = 'mymodel.pkl'
pickle.dump(mymodel, open(filename, 'wb'))
```
